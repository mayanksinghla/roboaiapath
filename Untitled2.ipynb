```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("interview.csv")
```

```python
df
```

```python
#First 5 Rows
print(df.head())
```

```python
#Data Types
print(df.dtypes)
```

```python
#Missing Values
print(df.isnull().sum())
```

```python
#Duplicate Rows
print(df.duplicated().sum())
```

```python
# Total Sales
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
```

```python
# Sales Region
sales_region = df.groupby("Region")["Sales"].sum()
print("\nSales by Region:")
print(sales_region)
```

```python
# Sales Product
sales_product = df.groupby("Product")["Sales"].sum()
print("Sales by Product:")
print(sales_product)
```

```python
# which product generated the most total sales?
sales_product.idxmax()
```

```python
# which region generated the most total sales?
sales_region.idxmax()
```

```python
sales_region.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```
